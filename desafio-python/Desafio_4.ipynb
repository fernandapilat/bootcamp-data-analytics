## **2.** Verificar os jogos com maior quantidade de gols na história da Bundesliga





**Criação da váriavel *TotalGols***
* **Passo 1:** nesta etapa optei por criar uma nova coluna com a soma das variáveis `AwayTeam` e `HomeGoals`.


```python
#Passo 1
df_fut['TotalGols'] = df_fut['HomeGoals'] + df_fut['AwayGoals']
df_fut.head(2)
```

**Explorar o dataframe para descobrir a maior quantidade de gols**

* **Passo 2:** utilizei a função `value_counts` na coluna TotalGols foi possível identificar que o maior número de gols no campeonato Bundesliga foi **12** e que esse placar ocorreu em **5** jogos.
* **Passo 3:** após classifiquei o dataframe pela coluna TotalGols com o intuito de localizar o id dos cinco primeiros jogos. Nessa fase, utilizei a função `sort_values` ao invés da função `max()`.

```python
#Passo 2
df_fut['TotalGols'].value_counts(ascending = True)
```

```python
#Passo 3
df_fut.sort_values(by='TotalGols', ascending=False)
```

**Filtro das informações dos jogos com mais gols**

* **Passo 4:** nesta fase, optei por utilizar a função `loc[]` para filtrar os cinco jogos com maior número de gols e criei um novo dataframe com todas as informações. Bem como, utilizei o método `reset_index` para remover o antigo índice, o objetivo é facilitar o `loc[]` para o print que informa ao locutor todos os jogos.



```python
#Passo 4
df_fut_top5 = df_fut.loc[[4457, 4052, 2456, 81, 5788]]
df_fut_top5.reset_index(drop=True, inplace=True)
df_fut_top5.head()
```

**Quais são os jogos com maior quantidade de gols?**

* **Passo 5:** construí o print que informa ao narrador os jogos com os maiores placares, para isso utilizei a função `loc[]` buscando diretamente no dataframe as infos, bem como a função `str()` para valores `int()`.




```python
#Passo 5
print('Os cinco jogos com maior placar são: ')
print()
print('(1): ' + df_fut_top5.loc[0, 'HomeTeam'] + ' vs. ' + df_fut_top5.loc[0, 'AwayTeam'] + '. Quantidade total de gols: ' 
      + str(df_fut_top5.loc[0, 'TotalGols']) + '.\n'
      '(2): ' + df_fut_top5.loc[1, 'HomeTeam'] + ' vs. ' + df_fut_top5.loc[1, 'AwayTeam'] + '. Quantidade total de gols: ' 
      + str(df_fut_top5.loc[1, 'TotalGols']) + '.\n'
      '(3): ' + df_fut_top5.loc[2, 'HomeTeam'] + ' vs. ' + df_fut_top5.loc[2, 'AwayTeam'] + '. Quantidade total de gols: ' 
      + str(df_fut_top5.loc[2, 'TotalGols']) + '.\n'
      '(4): ' + df_fut_top5.loc[3, 'HomeTeam'] + ' vs. ' + df_fut_top5.loc[3, 'AwayTeam'] + '. Quantidade total de gols: ' 
      + str(df_fut_top5.loc[3, 'TotalGols']) + '.\n'
      '(5): ' + df_fut_top5.loc[4, 'HomeTeam'] + ' vs. ' + df_fut_top5.loc[4, 'AwayTeam'] + '. Quantidade total de gols: ' 
      + str(df_fut_top5.loc[4, 'TotalGols']) + '.'
)
```

## **3.** Informar a data, os times participantes e o placar do jogo


 **Print das informações para o jornalista**

Nessa última etapa não fica claro se o narrador quer saber as informações referentes aos cinco jogos ou apenas um jogo. Como no **enunciado** fica claro que é **apenas um jogo**, conforme a frase: ***o narrador precisa informar o jogo com o maior número de gols na história da Bundesliga (Campeonato Alemão)***, decidi então utilizar as variáveis número de gols e quantidade de rounds, atendendo respectivamente ao maior número de gols e menor número de rounds.

* **Passo 6:** Primeiro fiz a alteração do formato da data para remover as horas e deixá-la mais no formato brasileiro. Utilizei a função `pd.DatetimeIndex` e `strftime('%d/%m/%Y')`.
* **Passo 7:** Por último, utilizei a função print para informar o jogo com maior número de gols e menor número de rounds.



```python
#Passo 6
df_fut_top5['Date'] = pd.DatetimeIndex(df_fut_top5['Date']).strftime('%d/%m/%Y')
df_fut_top5.head(5)
```

```python
#Passo 7
print('O jogo com o maior número de gols e menor round ocorreu no dia ' 
      + df_fut_top5.loc[3, 'Date'] + '\nentre os times ' + df_fut_top5.loc[3, 'HomeTeam'] + ' vs. ' 
      + df_fut_top5.loc[3, 'AwayTeam'] + ' com o total de ' + str(df_fut_top5.loc[3,'TotalGols']) 
      + ' gols e ' + str(df_fut_top5.loc[3, 'Round']) + ' rounds.')
```
