## **Desafio 2 - Conversão**

**Definição do problema:**

Dada a tabela com a média de altura e peso de mulheres estadunidenses, onde os nomes das colunas estão em inglês height(altura) e weight(peso) e seus respectivos valores estão nas seguintes unidades: polegada e libras. 

1. Construa uma nova representação onde os nomes das colunas estão em português, e os valores de altura são convertidos para centímetros e os valores de peso convertidos para quilograma. 

2. Plote o scatterplot de altura x peso da nova representação.

**Código do dataset:** Women

**Dicas:**

1 polegada = 2,54 centímetros

1 libra = 0,45 quilos

use o método .plot.scatter()

```python
#carregar dados
df_women = data('women')
df_women.head()
```

```python
#adicionar novas colunas com valores de centímetros e quilos
df_women['valor_altura'] = float(2.54)
df_women['valor_peso'] = float(0.45)

#converter altura e peso
df_women['altura'] = df_women['height'] * df_women['valor_altura']
df_women['peso'] = df_women['weight'] * df_women['valor_peso']

df_women.head(2)
```

```python
#filtrando/selecionando apenas as colunas altura e peso
df_women2 = df_women.loc[:,['altura', 'peso']]
df_women2.head(2)
```

```python
#plotando o gráfico de pontos
plt.scatter(df_women2['peso'], df_women2['altura'], color='#80B3FF', marker = 'o')

plt.ylabel('Altura (cm)')
plt.xlabel('Peso (kg)')
plt.title('Média de altura e peso de mulheres estadunidenses')
```

2. É possível fazer operações entre colunas: soma, subtração, divisão, multiplicação, etc.

## **Desafio 6 - Rendimento Ações**

**Definição do problema:** Você está trabalhando num banco e preciso informar ao seu cliente em que ano foi a maior rentabilidade de dividendo pago por ação da empresa Johnson&Jonson

1. Agregar os rendimentos de um mesmo ano

2. Calcular a valorização do rendimento de um ano para outro

3. Informar o ano de maior rentabilidade

**Código do dataset**: JohnsonJohnson

**Dicas:**

- Criar uma nova representação dos dados, onde a informação é anual

- Criar uma nova coluna informando a valorização anual

- numpy.trunc()

- método .shift()

- método .idxmax()

```python
#importando dados
df_jj = data('JohnsonJohnson')
df_jj.head(2)

#Visualizar a documentação do dataset
#data('JohnsonJohnson', show_doc = 'True')
```

```python
#criar um novo dataframe sem os valores trimestrais
df_jj_rend = {'ano': df_jj['time'].astype(int), 
              'rendimento': df_jj['JohnsonJohnson']}

#df_jj_rend = {'ano': df_jj['time'].replace('[^\d]', '', regex=True).astype(int), 
#              'rendimento': df_jj['JohnsonJohnson']}

df_jj2 = pd.DataFrame(df_jj_rend)
```

```python
#agregar os rendimentos de um mesmo ano
df_jj_var = df_jj2.groupby(['ano']).sum().reset_index()

#dataframe 1
df_jj_var.head(2)
```

```python
#excluir o valor do primeiro ano
df_jj_var2 = df_jj_var.drop(0).rename(columns={'ano': 'ano2',
                                               'rendimento': 'rendimento2'})

#dataframe 2
df_jj_var2.head(2)
```

```python
#resetar o indice

df_jj_var2.reset_index(drop=True, inplace=True)
df_jj_var2.head(2)
```

```python
#concatenar as duas colunas 
df_jj_final = pd.concat([df_jj_var, df_jj_var2], axis = 1)
```

```python
#tratando valores nulos no dataframe final
df_jj_final[['ano2', 'rendimento2']] = df_jj_final[['ano2', 'rendimento2']].fillna(method='ffill')
df_jj_final.head(2)
```

```python
#convertendo a coluna ano2 para número inteiro
df_jj_final['ano2'] = df_jj_final['ano2'].astype(int)
```

```python
#calcular a valorização do rendimento de um ano para outro
df_jj_final['var_rend_anual'] = round(((df_jj_final['rendimento2']/df_jj_final['rendimento'] * 100) - 100), 2)
df_jj_final.head(2)
```

```python
#classificar o ano de maior rentabilidade
df_jj_final.sort_values(by='var_rend_anual', ascending=False).head(1)
```

```python
#informar o ano de maior rentabilidade
print('O ano de maior rendimento foi ' + str(df_jj_final.loc[9, 'ano2']) + '.')
```

```python
plt.plot(df_jj_final['ano2'], df_jj_final['var_rend_anual'], color='darkturquoise', linestyle='-', linewidth = 2)

plt.ylabel('Rendimento')
plt.xlabel('Ano')
plt.title('Diferença anual dos rendimentos da J&J')
```

### **Aprendizados**

1. `replace('[^\d]', ...)` não resultou em nenhuma mudança, na verdade utilizando o `astype(int)` na coluna do ano consegui remover os valores após o ponto.
2. `pd.concat` é possível unir dois dataframes completamente diferentes e o que é mais legal é que você pode definir se você quer que a nova tabela seja adicionada abaixo utilizando o `axis=0` ou ao lado utilizando `axis=1`. Também é possível usar `merge`.
3. `fillna(method='ffill')` essa função é perfeita para tratar valores nulos. O método `ffill` me possibilitou copiar o resultado anterior na próxima linha e transformar todos os on dados anuais em inteiros.

Código do dataset: housing

**Dicas:**

- Uma medição da condição da pessoa é feita a cada seis meses.

**Linhas do Projeto/Estudo:**

* Nos 60 dias anteriores à entrevista inicial, **64% (232)** dos participantes passaram mais de uma semana nas ruas ou em abrigos.
* Os **36% restantes (130)** eram indivíduos que não possuíam um lar estável, ou que haviam passado um tempo extenso recentemente em uma instituição psiquiátrica.
* A previsão de que os participantes com acesso a certificados da Seção 8 entrariam em moradia independente estável mais rapidamente do que aqueles sem um certificado da Seção 8 também se aplicava aos participantes em ambas as condições que alcançaram algum tipo de moradia estável. Nessa análise, testamos a interação entre três variáveis: condição da Seção 8, tipo de moradia estável (independente vs. outra moradia comunitária) e o tempo para alcançar moradia estável (cedo vs. tarde). Um modelo log-linear testando a interação entre essas três variáveis apoiou fortemente a previsão de que os clientes com acesso a certificados estabilizariam em moradia independente mais rapidamente do que aqueles sem acesso a um certificado (LR(1) = 16,7, p < .01). Como é evidente na Tabela 6 (b), os clientes com acesso a certificados da Seção 8 tinham 8,4 vezes mais chances de alcançar moradia independente nos primeiros seis meses do que os clientes sem acesso a certificados da Seção 8. Por outro lado, os clientes sem acesso a certificados da Seção 8 tinham 3,4 vezes mais chances de alcançar outros tipos de moradia comunitária nos primeiros seis meses do que os clientes com acesso a certificados da Seção 8.

```python
#Passo 1 - Importando dados
df_moradia = data('housing')
df_moradia.head(5)
```

```python
#Passo 2 - Redefinir o nome das colunas
df_moradia.rename(columns={'id': 'id',
                           'y': 'status_moradia',
                           'time': 'tempo_sem',
                           'sec': 'zona'}, 
                  inplace=True)

df_moradia.head(2)
```
