### **Q1** - Em quantas medições, em média, as pessoas da base estava em situação de rua?

```python
#Passo 2 - Redefinir o nome das colunas
df_moradia.rename(columns={'id': 'id',
                           'y': 'status_moradia',
                           'time': 'tempo_sem',
                           'sec': 'zona'}, 
                  inplace=True)

df_moradia.head(2)
```

```python
#Passo 3 - Criar coluna com o valor percentual das ocorrências de valores por status de moradia 
df_moradia_perc = df_moradia.groupby(['status_moradia']).count()
df_moradia_perc2 = df_moradia_perc.reset_index().drop(columns=['tempo_sem', 'zona'])
df_moradia_perc2['perc'] = round((df_moradia_perc2['id'] / df_moradia_perc2['id'].sum()) * 100, 2)
```

```python
#Imprimir resposta

print('De acordo com as medições, em média, ' + str(df_moradia_perc2.loc[0, 'perc']) + 
      '% estavam em situação de rua.')
```

### **Q1** - Crie uma função que ao informar o sexo e a faixa etária de um suícidio é retornado as probabilidades do método utilizado

```python
#Carregar o dataframe com os dados de Suicídio

df_suicidios = data('Suicide')
df_suicidios.head(2)

#data('Suicide', show_doc = 'True')
```

```python
#Passo 1 - Mudar o nome das colunas
df_suicidios.rename(columns={'Freq': 'freq',
                             'sex': 'genero',
                             'method': 'metodo1',
                             'age': 'idade',
                             'age.group': 'faixa_etaria',
                             'method2': 'metodo2'},
                     inplace=True)

df_suicidios.head(2)
```

```python
#Passo 2 - Ajustar 

#Traduzir o gênero para o português
df_suicidios['genero'] = df_suicidios['genero'].replace('male', 'masculino')
df_suicidios['genero'] = df_suicidios['genero'].replace('female', 'feminino')

#Traduzir os métodos 1 para o português
df_suicidios['metodo1'] = df_suicidios['metodo1'].replace('poison', 'veneno')
df_suicidios['metodo1'] = df_suicidios['metodo1'].replace('cookgas', 'gás de cozinha')
df_suicidios['metodo1'] = df_suicidios['metodo1'].replace('toxicgas', 'gás tóxico')
df_suicidios['metodo1'] = df_suicidios['metodo1'].replace('hang', 'enforcar')
df_suicidios['metodo1'] = df_suicidios['metodo1'].replace('drown', 'afogar')
df_suicidios['metodo1'] = df_suicidios['metodo1'].replace('gun', 'arma de fogo')
df_suicidios['metodo1'] = df_suicidios['metodo1'].replace('knife', 'faca')
df_suicidios['metodo1'] = df_suicidios['metodo1'].replace('jump', 'pular')
df_suicidios['metodo1'] = df_suicidios['metodo1'].replace('other', 'outro')


#Traduzir os métodos 2 para o português
df_suicidios['metodo2'] = df_suicidios['metodo2'].replace('poison', 'veneno')
df_suicidios['metodo2'] = df_suicidios['metodo2'].replace('gas', 'gás')
df_suicidios['metodo2'] = df_suicidios['metodo2'].replace('toxicgas', 'gás tóxico')
df_suicidios['metodo2'] = df_suicidios['metodo2'].replace('hang', 'enforcar')
df_suicidios['metodo2'] = df_suicidios['metodo2'].replace('drown', 'afogar')
df_suicidios['metodo2'] = df_suicidios['metodo2'].replace('gun', 'arma de fogo')
df_suicidios['metodo2'] = df_suicidios['metodo2'].replace('knife', 'faca')
df_suicidios['metodo2'] = df_suicidios['metodo2'].replace('jump', 'pular')
df_suicidios['metodo2'] = df_suicidios['metodo2'].replace('other', 'outro')

df_suicidios.head(2)
```

```python
#Agrupar pelo genero e depois faixa etária, por fim criar uma coluna com a probabilidade
df_prob = df_suicidios.groupby(['genero', 'faixa_etaria', 'metodo1'])['freq'].sum().to_frame().reset_index()
df_prob['prob'] = (df_prob['freq'] / df_prob['freq'].sum() * 100).round(2)
df_prob.head(2)
```
